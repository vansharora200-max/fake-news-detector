```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size= 0.2, random_state= 42, stratify= Y
)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(Y_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    max_features = 5000,
    ngram_range = (1,2),
    min_df = 2,
    max_df = 0.95
)

vectorizer.fit(X_train)
```

```python
len(vectorizer.vocabulary_)
```

```python
sorted(vectorizer.vocabulary_.keys())[:20]
```

```python
vectorizer.vocabulary_
```

```python
X_train_tfidf = vectorizer.transform(X_train)
```

```python
X_test_tfidf = vectorizer.transform(X_test)
```

```python
X_train_tfidf.shape
```

```python
X_test_tfidf.shape
```

```python
total_values = X_train_tfidf.shape[0] * X_train_tfidf.shape[1]
total_values
```

```python
nonzero_values = X_train_tfidf.nnz
```

Sparsity(%)

```python
round((1 - nonzero_values/total_values) * 100, 2)
```

TF-IDF scores for the first article

```python
import numpy as np

first_row = X_train_tfidf[0]
nonzero_indices = first_row.nonzero()[1]
feature_names = vectorizer.get_feature_names_out()
```

Number of non zero features first training article has 

```python
len(nonzero_indices)
```

```python
scores = [(feature_names[i], round(first_row[0, i], 4)) for i in nonzero_indices]
scores_sorted = sorted(scores, key= lambda x: x[1], reverse = True)[:10]
for term, score in scores_sorted:
    print(f" {term:30s} {score}")
```
