# Text Cleaning

```python
import re
```

```python
def clean(text):
    text = text.lower()
    text = re.sub(r'http\S+ | www\S+', '', text)
    text = re.sub(r'@\w+ | #\w+', '', text)
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = ' '.join([i for i in text.split() if len(i) > 1])
    text = re.sub(r'\s+', ' ', text)
    return text
```

### Updating the clean() function to remove stopwords in the text as well.

```python
def clean(text):
    text = text.lower()
    text = re.sub(r'http\S+ | www\S+', '', text)
    text = re.sub(r'@\w+ | #\w+', '', text)
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = ' '.join([i for i in text.split() if i not in stop_words and len(i) > 1]) # Updated this line
    text = re.sub(r'\s+', ' ', text)
    return text
```

```python
df['Cleaned'] = df['Content'].apply(clean)
```

### Average word count before and after cleaning of text, and reduction of word count on avg. basis.

```python
df['original_wc'] = df['Content'].apply(lambda x: len(x.split()))
df['cleaned_wc'] = df['Cleaned'].apply(lambda x: len(x.split()))
```

```python
a = float(f"{df['original_wc'].mean():.1f}")
a
```

```python
b = float(f"{df['cleaned_wc'].mean():.1f}")
b
```
